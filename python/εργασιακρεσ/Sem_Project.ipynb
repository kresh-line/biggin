## Άσκηση 1: Έλεγχος Αναγραμματισμού
**Ζητούμενο:** Να γραφτεί μια συνάρτηση που ελέγχει αν δύο συμβολοσειρές είναι αναγραμματισμοί, δηλαδή αν αποτελούνται από ακριβώς τους ίδιους χαρακτήρες με διαφορετική σειρά.

**Είσοδος**: word1 = "listen", word2 = "silent"

**Έξοδος**: Is "listen" an anagram of "silent"? True

```python
def is_anagram(word1, word2):
    # η sorted() ταξινομεί τους χαρακτήρες - αν είναι ίδιοι, είναι αναγραμματισμός
    return sorted(word1) == sorted(word2)

word1 = "listen"
word2 = "silent"

result = is_anagram(word1, word2)
print('Is "' + word1 + '" an anagram of "' + word2 + '"?', result)
```

## Άσκηση 2: Αντιστροφή Κάθε Λέξης Μιας Συμβολοσειράς
**Ζητούμενο:** Να γραφτεί μια συνάρτηση που, δεδομένης μιας πρότασης, αντιστρέφει κάθε λέξη ξεχωριστά, ενώ διατηρεί τη σειρά των λέξεων όπως εμφανίζονται αρχικά στο κείμενο.

**Είσοδος**: "Python is awesome"

**Έξοδος**: "nohtyP si emosewa"

```python
def reverse_words(sentence):
    words = sentence.split()  # χωρίζει σε λέξεις
    result = ""
    for w in words:
        result += w[::-1] + " "  # αντιστρέφει κάθε λέξη με slicing
    return result.strip()

sentence = "Python is awesome"
print(reverse_words(sentence))
```

## Άσκηση 3: Φιλτράρισμα με List Comprehension
**Ζητούμενο:** Να χρησιμοποιηθεί μία μόνο έκφραση list comprehension για να εξαχθούν από μια λίστα συμβολοσειρών εκείνες που πληρούν και τις δύο από τις ακόλουθες προϋποθέσεις:

1) Έχουν μήκος μεγαλύτερο από 5 χαρακτήρες.

2) Ξεκινούν με φωνήεν (a, e, i, o, u).

**Είσοδος**: ["apple", "education", "ice", "ocean", "python", "umbrella"]

**Έξοδος**: ['education', 'umbrella']

```python
# φωνήεντα για τον έλεγχο αρχικού γράμματος
vowels = "aeiou"

words = ["apple", "education", "ice", "ocean", "python", "umbrella"]

# list comprehension: κρατάμε λέξεις με μήκος > 5 που αρχίζουν με φωνήεν
result = [w for w in words if len(w) > 5 and w[0].lower() in vowels]

print(result)
```

## Άσκηση 4: Συμμετρική Διαφορά Συνόλων
**Ζητούμενο:** Σε 2 λίστες να βρεθούν, τα στοιχεία που εμφανίζονται είτε στην πρώτη είτε στη δεύτερη λίστα, αλλά όχι και στις δύο ταυτόχρονα.

**Είσοδος**: list1 = [101, 102, 103], list2 = [103, 104, 105]

**Έξοδος**: {101, 102, 104, 105}

```python
list1 = [101, 102, 103]
list2 = [103, 104, 105]

# μετατρέπουμε τις λίστες σε σύνολα και βρίσκουμε τη συμμετρική διαφορά
# ο τελεστής ^ επιστρέφει στοιχεία που υπάρχουν σε ένα μόνο από τα δύο σύνολα
result = set(list1) ^ set(list2)

print(result)
```

## Άσκηση 5: Εύρεση Διπλότυπων
**Ζητούμενο:** Να γραφτεί μια συνάρτηση που εντοπίζει όλα τα διπλότυπα στοιχεία μέσα σε μια λίστα.

**Είσοδος**: numbers = [1, 2, 3, 2, 4, 5, 1, 6]

**Έξοδος**: Duplicates found: {1, 2}

```python
def find_duplicates(numbers):
    seen = set()       # σύνολο με τα στοιχεία που έχουμε δει
    duplicates = set() # σύνολο με τα στοιχεία που εμφανίζονται παραπάνω από μία φορά

    for n in numbers:
        if n in seen:
            duplicates.add(n)  # αν το έχουμε ξαναδεί, είναι διπλότυπο
        else:
            seen.add(n)        # πρώτη εμφάνιση, το αποθηκεύουμε

    return duplicates

numbers = [1, 2, 3, 2, 4, 5, 1, 6]
print("Duplicates found:", find_duplicates(numbers))
```

## Άσκηση 6: Έλεγχος Εγκυρότητας Κωδικού Πρόσβασης
**Ζητούμενο:** Να υλοποιηθεί μια συνάρτηση που ελέγχει αν ένας κωδικός πρόσβασης πληροί συγκεκριμένα κριτήρια ασφαλείας. Η συνάρτηση πρέπει να επιστρέφει αν ο κωδικός είναι έγκυρος, καθώς και λίστα με όλα τα μηνύματα σφάλματος που αντιστοιχούν στους κανόνες που δεν ικανοποιούνται.
Ο κωδικός θεωρείται έγκυρος μόνο αν ικανοποιεί και τους τέσσερεις παρακάτω κανόνες:

1) Μήκος τουλάχιστον 8 χαρακτήρες.
2) Να περιέχει τουλάχιστον έναν αριθμό.
3) Να περιέχει τουλάχιστον ένα κεφαλαίο γράμμα.
4) Να μην περιέχει κενά (space).

Αν κάποιος κανόνας δεν τηρείται, προστίθεται αντίστοιχο μήνυμα σε μια λίστα σφαλμάτων.

Η συνάρτηση πρέπει να επιστρέφει ένα ζεύγος (tuple), το οποίο θα περιέχει:

- True/False ανάλογα με το αν ο κωδικός είναι έγκυρος
- Τη λίστα με τα σφάλματα που εντοπίστηκαν

**Είσοδος**: 
1) Test1234 
2) we ak

**Έξοδος**: 
1) (True, [])
2) (False, ['Πολύ μικρός κωδικός', 'Δεν περιέχει αριθμό', 'Δεν περιέχει κεφαλαίο γράμμα', 'Περιέχει κενό'])

**Hint:** Προτείνεται η χρήση των συναρτήσεων isdigit(), isupper() και isspace()

```python
def check_password(password):
    errors = []  # λίστα που συλλέγει τα μηνύματα σφάλματος

    # κανόνας 1: ελάχιστο μήκος 8 χαρακτήρων
    if len(password) < 8:
        errors.append("Πολύ μικρός κωδικός")

    # κανόνας 2: πρέπει να περιέχει τουλάχιστον έναν αριθμό
    has_digit = False
    for c in password:
        if c.isdigit():
            has_digit = True
    if not has_digit:
        errors.append("Δεν περιέχει αριθμό")

    # κανόνας 3: πρέπει να περιέχει τουλάχιστον ένα κεφαλαίο γράμμα
    has_upper = False
    for c in password:
        if c.isupper():
            has_upper = True
    if not has_upper:
        errors.append("Δεν περιέχει κεφαλαίο γράμμα")

    # κανόνας 4: δεν πρέπει να περιέχει κενό
    for c in password:
        if c.isspace():
            errors.append("Περιέχει κενό")
            break  # αρκεί ένα κενό για να προστεθεί το σφάλμα

    # επιστρέφει tuple: (True/False, λίστα σφαλμάτων)
    return (len(errors) == 0, errors)

print(check_password("Test1234"))
print(check_password("we ak"))
print(check_password("weakpassword"))
```

## Άσκηση 7: Παραγωγός Fibonacci
**Ζητούμενο:** Να δημιουργηθεί μια συνάρτηση που να παράγει διαδοχικούς αριθμούς Fibonacci μέχρι το N.
**Είσοδος**: n = 8

**Έξοδος**: First 8 Fibonacci numbers:
0 1 1 2 3 5 8 13 


```python
def fibonacci(n):
    a, b = 0, 1  # αρχικές τιμές: F(0)=0, F(1)=1
    for _ in range(n):
        yield a          # επιστρέφει την τρέχουσα τιμή χωρίς να σταματά η συνάρτηση
        a, b = b, a + b  # η επόμενη τιμή είναι το άθροισμα των δύο προηγούμενων

n = 8
print(f"First {n} Fibonacci numbers:")
for num in fibonacci(n):
    print(num, end=" ")
```

## Άσκηση 8: Εύρεση Μεγαλύτερου και Μικρότερου Ψηφίου
**Ζητούμενο:** Να γραφτεί ένα πρόγραμμα που εντοπίζει το μεγαλύτερο και το μικρότερο ψηφίο μέσα σε έναν ακέραιο αριθμό.

**Είσοδος**: num = 75869

**Έξοδος**: 

Largest digit: 9

Smallest digit: 5


```python
def max_min_digit(num):
    # μετατρέπουμε τον αριθμό σε λίστα ψηφίων (ακέραιοι)
    digits = [int(d) for d in str(num)]

    print("Largest digit:", max(digits))   # το μεγαλύτερο ψηφίο
    print("Smallest digit:", min(digits))  # το μικρότερο ψηφίο

num = 75869
max_min_digit(num)
```

## Άσκηση 9: Έλεγχος Αριθμού Armstrong
**Ζητούμενο:** Να γραφτεί ένα πρόγραμμα που ελέγχει αν ένας αριθμός είναι αριθμός Armstrong. Ένας αριθμός θεωρείται Armstrong όταν το άθροισμα των ψηφίων του, υψωμένων στη δύναμη του πλήθους των ψηφίων του, ισούται με τον ίδιο τον αριθμό.

Για έναν τριψήφιο αριθμό, ένας αριθμός θεωρείται Armstrong όταν το άθροισμα των κύβων των ψηφίων του είναι ίσο με τον ίδιο τον αριθμό.
πχ, 153 = 1^3 + 5^3 + 3^3.


**Είσοδος**: num = 153

**Έξοδος**: 153 IS an Armstrong number


***Bonus (+1 στην τελικη βαθμολογία)***: Να εμφανίζει αναλυτικά τα βήματα του υπολογισμού, με μορφή τύπου:
153 = 1³ + 5³ + 3³

**Είσοδος**: num = 153

**Έξοδος**: 

Steps for number 153:\
1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153\
153 IS an Armstrong number.

```python
def is_armstrong(num):
    digits = str(num)   # μετατρέπουμε τον αριθμό σε string για να πάρουμε τα ψηφία
    n = len(digits)     # πλήθος ψηφίων (η δύναμη που θα χρησιμοποιήσουμε)

    # υπολογισμός αθροίσματος: κάθε ψηφίο υψωμένο στη δύναμη n
    total = sum(int(d) ** n for d in digits)

    # εκτύπωση αναλυτικών βημάτων (bonus)
    steps = " + ".join(f"{d}^{n}" for d in digits)
    values = " + ".join(str(int(d) ** n) for d in digits)
    print(f"Steps for number {num}:")
    print(f"{steps} = {values} = {total}")

    # σύγκριση αθροίσματος με τον αρχικό αριθμό
    if total == num:
        print(f"{num} IS an Armstrong number.")
    else:
        print(f"{num} is NOT an Armstrong number.")

num = 153
is_armstrong(num)
```

## Άσκηση 10: Εύρεση Όλων Των Αριθμών Armstrong Μέχρι Ένα Αριθμό
**Ζητούμενο:** Να γραφτεί ένα πρόγραμμα που, δεδομένου ενός ακέραιου αριθμού Ν, επιστρέφει μια λίστα με όλους τους αριθμούς Armstrong από το 1 μέχρι το Ν.

**Είσοδος**: num = 5000

**Έξοδος**: [1, 2, 3, 4, 5, 6, 7, 8, 9, 153, 370, 371, 407, 1634]

```python
def is_armstrong_check(num):
    digits = str(num)
    n = len(digits)
    # ελέγχει αν το άθροισμα των ψηφίων υψωμένα στη δύναμη n ισούται με τον αριθμό
    return sum(int(d) ** n for d in digits) == num

def all_armstrong(N):
    # επιστρέφει λίστα με όλους τους Armstrong από 1 έως N
    return [i for i in range(1, N + 1) if is_armstrong_check(i)]

num = 5000
print(all_armstrong(num))
```
