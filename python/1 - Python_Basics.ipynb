### Είδη Μεταβλητών στην Python
Tα βασικά είδη μεταβλητών (data types) στην Python.

#### 1. Ακέραιοι (int)
Ο τύπος `int` χρησιμοποιείται για ακέραιους αριθμούς.

````python
x = 10
type(x)
````

#### 2. Δεκαδικοί (float)
Ο τύπος `float` αναπαριστά αριθμούς με δεκαδικά.

````python
y = 3.14
type(y)
````

#### 3. Αλφαριθμητικά (str)
Ο τύπος `str` χρησιμοποιείται για κείμενο.

````python
name = 'Python'
type(name)
````

#### 4. Boolean (bool)
Ο τύπος `bool` παίρνει τιμές `True` ή `False`.

````python
flag = True
type(flag)
````

#### 5. Λίστες (list)
Οι λίστες μπορούν να περιέχουν πολλά στοιχεία διαφόρων τύπων.

````python
lst = [1, 2, 3, 'a']
type(lst)
````

#### 6. Tuples
Τα tuples μοιάζουν με λίστες αλλά είναι μη μεταβαλλόμενες (immutable).

````python
t = (1, 2, 3)
type(t)
````

#### 7. Σύνολα (set)
Τα σύνολα περιέχουν μοναδικές τιμές χωρίς σειρά.

````python
s = {1, 2, 3, 3}
type(s)
````

#### 8. Λεξικά (dict)
Τα λεξικά αποθηκεύουν ζεύγη κλειδιού-τιμής.

````python
d = {'name': 'Athanasios', 'age': 30}
type(d)
````
