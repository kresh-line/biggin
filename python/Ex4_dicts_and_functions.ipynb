## Άσκηση 19
**Παιχνίδι Μάντεψια Αριθμού**

Αναπτύξτε μια συνάρτηση για ένα παιχνίδι μάντεψιάς αριθμού, όπου το όρισμα ειναι ένας τυχαίος αριθμός (1-100) και κάποιος άλλος χρηστης προσπαθεί να τον μαντέψει.
Μετά από κάθε προσπάθεια εμφανίζει: `Πολύ μικρός`, `Πολύ μεγάλος`, ή `Μπράβο!`.
Να επιτρέπονται έως 10 προσπάθειες και να εμφανίζεται το πλήθος προσπαθειών στο τέλος.
