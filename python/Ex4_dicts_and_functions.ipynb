Δημιουργήστε μια συνάρτηση η οποία δέχεται σαν όρισμα έναν ακέραιο N και εμφανίζει όλους τους αριθμούς από το 1 έως το N στην ίδια γραμμή.

## Άσκηση 2
Υλοποιήστε μια συνάρτηση που υπολογίζει και επιστρέφει το άθροισμα των αριθμών από το 1 έως το $N$ χρησιμοποιώντας αποκλειστικά τη δομή επανάληψης while

```python
def sum(N):
    s = 0
    i = 1
    while i <= N:
        s += i
        i += 1
    print("Άθροισμα:", s)

sum(6)
```

## Άσκηση 3
Δημιουργήστε μια συνάρτηση που δέχεται μια λίστα ακεραίων και επιστρέφει ένα λεξικό (dictionary) με το πλήθος των άρτιων και των περιττών στοιχείων.

πχ [3, 10, 5, 8, 12, 7, 6] #Σωστη απαντηση {'αρτια':4, 'περιττα': 3}

```python
def arrr (N):
    even = 0   
    odd = 0   
    for x in N:
        if x % 2 == 0:  
            even += 1
        else:
            odd += 1
    print("Άρτια:", even)
    print("Περιττά:", odd)
arrr ([3, 10, 5, 8, 12, 7, 6])
```

## Άσκηση 4
Δημιουργήστε μια συνάρτηση που να υπολόγιζει το παραγοντικό `n!` ( n! = 1⋅2⋅3⋅ …⋅(n-1) ⋅n) με `for`.

```python
def fal (N):
    fact = 1

    for i in range(1, N + 1):
        fact *= i
    return fact 

print("Το παραγοντικό είναι:", fal(5))

```

## Άσκηση 6
Υλοποιήστε μια συνάρτηση που δέχεται αγγλικό κείμενο και επιστρέφει ένα λεξικό με το πλήθος φωνηέντων και συμφώνων, χρησιμοποιώντας ένα σύνολο (set) για τον εντοπισμό των φωνηέντων.

'Hint': Χρησιμοποιήστε την συνάρτηση isalpha() για καλύτερα αποτελέσματα

```python
def kfs (k):
    
    f = 0
    s = 0
    
    for x in k:
        if x.isalpha():
            if x.lower() in "aeiou":
                f += 1
            else:
                s += 1
    return f,s    
k = input("Δώσε κείμενο: ")
f, s = kfs(k)
print("Φωνήεντα:", f)
print("Σύμφωνα:", s)
```

## Άσκηση 8
Γράψτε μια συνάρτηση που χρησιμοποιεί τη δομή while για να βρει και να επιστρέψει το άθροισμα των ψηφίων ενός θετικού ακεραίου.

```python
def ns(n):
    
    s = 0
    
    while n > 0:
        s = s + n % 10
        n = n // 10
    return s
num = 53
print(ns(num))

    
```

## Άσκηση 9
Υλοποιήστε μια συνάρτηση που υπολογίζει τον Μέγιστο Κοινό Διαιρέτη (ΜΚΔ) δύο αριθμών με while.

```python
def AB(a,b):


    while a != b:
        if a > b:
            a = a - b
        else:
            b = b - a
    return a    
x = 12
y = 18

print(AB(x, y))
```
