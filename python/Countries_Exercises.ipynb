## 1) Να διαβάστει το αρχείο countries.csv και να αποθηκεύτούν τα δεδομένα σε λεξικό.

```python
with open('countries.csv') as f:
    data = f.read()


data = data.split('\n')


data = [x.split(',') for x in data]

data_clean=[]
for row in data:
    if"" not in row:
        data_clean.append(row)
data_clean = [[y.strip() for y in x] for x in data_clean]

```

```python
headers = data_clean[0]
headers
```

```python
headers.index('Country')
```

```python
countries=data_clean[1:]
```

```python
for x in countries:
    print(x [headers.index('Capital')])
```

```python
contry_list=[]

for cou in countries:
    contry_list.append({'Country':cou[headers.index('Country')],'Population':int(cou[headers.index('Population')]),'Area':float(cou[headers.index('Area')]),'Capital':cou[headers.index('Capital')],'Continent':cou[headers.index('Continent')]})
```
