# Λογικές μεταβλητές
Oι [λογικές μεταβλητές](https://el.wikipedia.org/wiki/%CE%86%CE%BB%CE%B3%CE%B5%CE%B2%CF%81%CE%B1_%CE%9C%CF%80%CE%BF%CF%85%CE%BB). Οι τιμές που μπορούν να πάρουν είναι: ```True``` ή ```False``` (προσοχή πρέπει να είναι κεφαλαίο το πρώτο γράμμα).

````python
a = True
a = False
````

Η πράξη ```and``` είναι ```True``` αν και οι δύo μεταβλητές είναι ```True```:

````python
True and True
````

````python
True and False
````

````python
False and True
````

````python
False and False
````

Η πράξη ```or``` είναι ```True``` αν έστω μία από τις δύο μεταβλητές είναι ```True```:

````python
True or True
````

````python
True or False
````

````python
False or True
````

````python
False or False
````

H πράξη ```not``` αντιστρέφει μία λογική έκφραση:

````python
not False
````

````python
not True
````

Μπορούμε να συνθέτουμε πολύπλοκες λογικές εκφράσεις:

````python
False and (True or False)
````

````python
True and (True or False)
````

To ```^``` ονομάζεται [bitwise XOR](https://en.wikipedia.org/wiki/Bitwise_operation#XOR) 

````python
True^False
````

Επίσης είναι και οι λογικοί τελεστές στην Python:
- `==` Έλεγχος ισότητας
- `>` Μεγαλύτερο
- `<` Μικρότερο
- `>=` Μεγαλύτερο ή ίσο
- `<=` Μικρότερο ή ίσο
- `!=` Διάφορο


````python
# == Έλεγχος ισότητας
a = 5
b = 5
a == b
````

````python
# > Μεγαλύτερο
a = 10
b = 3
a > b
````

````python
# < Μικρότερο
a = 2
b = 8
a < b
````

````python
1<3
````

````python
1<0
````

````python
0.0000000001 > 0
````

````python
# >= Μεγαλύτερο ή ίσο
a = 5
b = 5
a >= b
````

````python
# <= Μικρότερο ή ίσο
a = 4
b = 6
a <= b
````

````python
# != Διάφορο
a = 7
b = 10
a != b
````

Μπορούμε να συγκρίνουμε αλφαριθμητικά. Τα strings που είναι πιο "πάνω" στη κατάταξη θεωρούνται μικρότερα. Τα κεφαλαία είναι "μικρότερα" από τα μικρά. Οι αριθμοί είναι μικρότεροι από τα κεφαλαία.  

````python
"a" < "b"
````

````python
"a" < "ab"
````

````python
"A" < "a"
````

````python
"05456745674" < "5"
````

````python
"a" in "hello"
````

````python
"h" in "hello"
````

````python
"li" in "hello"
````

````python
"lo" in "hello"
````

````python
1 in [2,3,4]
````

````python
2 in [2,3,4]
````

````python
[] in [3,4,[],3]
````
