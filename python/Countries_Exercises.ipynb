# 📘 Άσκηση Ανάλυσης Δεδομένων Χωρών

## 1) Να διαβάστει το αρχείο countries.csv και να αποθηκεύτούν τα δεδομένα σε λεξικό.

## 2) Να εμφανίσετε τις πρώτες 5 εγγραφές.

## 3) Να προσθέσετε σε κάθε εγγραφή ένα νέο πεδίο που θα υπολογιζει την πυκνότητα (density = population / area) για κάθε χώρα.

## 4) Να εμφανίσετε όλες τις χώρες με πυκνότητα μεγαλύτερη από 200.

## 5) Να βρείτε τη χώρα με τη μεγαλύτερη πυκνότητα πληθυσμού.

## 6) Να υπολογίσετε τον μέσο πληθυσμό ανά ήπειρο.

## 7) Nα γράψετε μια συνάρτηση που θα δέχεται σαν ορίσματα το dataset και το όνομα μιας χώρας και θα έχει σαν έξοδο την πρωτεύουσά της αν υπάρχει.

## 8) Να εμφανίσετε τις 3 χώρες της Ευρώπης με τον μεγαλύτερο πληθυσμό.
