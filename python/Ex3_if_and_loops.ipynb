# Ασκήσεις if / for / while

## Άσκηση 1
Να εμφανίζονται οι αριθμοί από το 1 έως το N.

```python
N=5

for i in range(1, N + 1):
    print(i)
```

## Άσκηση 2
Να υπολογιστεί το άθροισμα 1..N χρησιμοποιώντας `while`.

```python
N = 5
sum = 0
i = 1

while i <= N:
    sum+= i
    i += 1

print("Άθροισμα:", sum)
```

## Άσκηση 3
Σε μια λίστα ακεραίων, μέτρησε πόσα στοιχεία είναι άρτια και πόσα περιττά.

πχ [3, 10, 5, 8, 12, 7, 6] #Σωστη απαντηση 4 αρτια 3 περιττα

```python
arr = [3, 10, 5, 8, 12, 7, 6]

even = 0   
odd = 0   

for x in arr:
    if x % 2 == 0:  
        even += 1
    else:
        odd += 1

print("Άρτια:", even)
print("Περιττά:", odd)
```

## Άσκηση 4
Υπολόγισε το παραγοντικό `n!` ( n! = 1⋅2⋅3⋅ …⋅(n-1) ⋅n) με `for`.

```python
n = 5
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Το παραγοντικό είναι:", fact)
```
